```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step (Thresholding) function

```python
def step(x):
    if x < 0:
        return 0
    else:
        return 1
```

```python
x = np.linspace(-10, 10, 100)
y = [step(i) for i in x]

plt.plot(x, y)
plt.grid(True)
plt.show()
```

## Sigmoid function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
y = [sigmoid(i) for i in x]

plt.plot(x, y)
plt.grid(True)
plt.show()
```

## Tanh function

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10, 100)
y = [tanh(i) for i in x]

plt.plot(x, y)
plt.grid(True)
plt.show()
```

## ReLU function 

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.linspace(-10, 10, 100)
y = [relu(i) for i in x]

plt.plot(x, y)
plt.grid(True)
plt.show()
```

## Leaky_ReLU function

```python
def leaky_relu(x, alpha):
    return np.where(x > 0, x, alpha*x)
```

```python
x = np.linspace(-10, 10, 100)
y = [leaky_relu(i, 0.1) for i in x]

plt.plot(x, y)
plt.grid(True)
plt.show()
```

## Softmax function

```python
def softmax(h):
    sumh = sum(np.exp(h))
    return np.exp(h)/sumh
```

```python
h = [9, 6, -2]
y = softmax(h)
print(y)
```
