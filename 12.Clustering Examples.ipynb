## Divide the following dataset in to 2 clusters.


```python
# Loading dataset
import numpy as np
import pandas as pd
X = pd.DataFrame({"Height":[5.2, 5.97, 5.18, 6.05, 5.12, 5.89], "Weight":[59.02, 84.01, 54.43, 91.25, 68.04, 84.84]})
print(X)
```

```python
# Building clustering model
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 2)
km.fit(X)
```

```python
# Displaying the cluster assignments and cluster centeroids (means)
print("Cluster Assignments:", km.labels_)
print("--------------------------")
print("Cluster centroids:", km.cluster_centers_)
```

```python
# Predict cluster assignment of the sample (5.8, 91) 
data = np.array([[5.8, 91]])
print("The sample assigned to cluster", km.predict(data)[0])
```
