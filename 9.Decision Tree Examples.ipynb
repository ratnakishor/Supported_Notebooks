## Example 1

```python
import numpy as np
import pandas as pd

X = pd.DataFrame({"X":[1,1,0,1], "Y":[1,1,0,0], "Z":[1,0,1,0]})
y = pd.Series(["A", "A", "B", "B"])

print("Training Data Input:")
print(X)
print("--------------")
print("Training Data Output:")
print(y)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion  = "entropy")
tree.fit(X,y)
```

```python
from sklearn.tree import plot_tree

plot_tree(tree, class_names = ["A", "B"], feature_names = ["X", "Y", "Z"], impurity = False, filled=True)
```

## Example 2

```python
import numpy as np
import pandas as pd

X = pd.DataFrame({"Height":["short", "tall", "tall", "tall", "short", "tall", "tall", "short"], 
                  "Hair":["blond", "blond", "red", "dark", "dark", "dark", "blond", "blond"],
                  "Eyes":["blue", "brown", "blue", "brown", "blue", "blue", "blue", "brown"]})
y = pd.Series(["yes", "no", "yes", "no", "no", "no", "yes", "no"])

print("Training Data Input")
print(X)
print("--------------")
print("Training Data Output")
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
X["Height"] = encoder.fit_transform(X["Height"])
X["Hair"] = encoder.fit_transform(X["Hair"])
X["Eyes"] = encoder.fit_transform(X["Eyes"])

print("Training Data Input:")
print(X)
print("--------------")
print("Training Data Output:")
print(y)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion  = "entropy")
tree.fit(X,y)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 10))
plot_tree(tree, class_names = ["yes", "no"], feature_names = ["Height", "Hair", "Eyes"], impurity = False, filled=True)
```
